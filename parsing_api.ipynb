```python
import requests
import os
import json
import pandas as pd
from datetime import datetime
from dotenv import load_dotenv
load_dotenv()
```

```python
API_TOKEN = os.getenv('token_weather_hourly')
```

```python
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36;"
}
params = {
    "q": "Запоріжжя",
    "appid": API_TOKEN,
    "units": "metric"}

url = 'https://api.openweathermap.org/data/2.5/forecast?'#q={city name}&appid={open_wheather_tok_hourly}'
```

```python
s = requests.Session()
s.headers.update(headers)
response = s.post(url=url,params=params).json()
```

```python
lst_3hour_forecast = []

for hours in response['list']:
    hour3 = {}
    hour3['date'] = str(datetime.fromtimestamp(hours['dt']))
    hour3['temperature'] = hours['main']['temp']
    hour3['temp feels like'] = hours['main']['feels_like']
    hour3['pressure'] = hours['main']['pressure']
    hour3['humidity'] = hours['main']['humidity']
    hour3['cloudy'] = hours['weather'][0]['description']
    hour3['cloudiness'] = hours['clouds']['all']
    hour3['wind'] = hours['wind']['speed']
    hour3['visibility'] = hours['visibility']
    hour3['probability'] = int(round((hours['pop'] * 100),3))
    if hours.get('snow'):
        hour3['snow'] = hours['snow'].get('3h','')
    else:
        hour3['snow'] = ''
    if hours.get('rain'):
        hour3['rain'] = hours['rain'].get('3h','')
    else:
        hour3['rain'] = ''
    lst_3hour_forecast.append(hour3)
```

```python
dataPD = pd.DataFrame(lst_3hour_forecast)
dataPD.to_csv(r'E:\111\forecast5d3H.csv', index = None)
dataPD.to_excel(r'E:\111\forecast5d3H.xlsx', index = None)
```
